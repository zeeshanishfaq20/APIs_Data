```python
import requests
import pandas as pd
from datetime import datetime

# API URL
url = 'https://api.sportradar.com/soccer-extended/trial/v4/en/competitors/sr%3Acompetitor%3A44/summaries.json?api_key=5qKS39A3oeSHA0HzXhlRXs89ciYKKHkiZEwWKEos'

# Make the request
r = requests.get(url)
data = r.json()

# Extract relevant data
summaries = data.get('summaries', [])

# Initialize lists to store extracted data
competitors_data = []
sport_event_status_data = []
statistics_data = []
statistics_oppenents_data = []

# Loop through each summary
for summary in summaries:

  # Extract competitors
  competitors = summary['sport_event']['competitors']

  for competitor in competitors:
    if competitor['name'] != 'Liverpool FC':
       timestamp = summary['sport_event']['start_time']
       date_only = datetime.strptime(timestamp, '%Y-%m-%dT%H:%M:%S%z').date()
       competitors_data.append({
          'competitor_id': competitor['id'],
          'name': competitor['name'],
          'country': competitor.get('country', None),
          'country_code': competitor.get('country_code', None),
          'abbreviation': competitor.get('abbreviation', None),
          'match_id': summary['sport_event']['sport_event_context']['competition']['id'],
          'match_played': date_only,
          'qualifier': competitor.get('qualifier', None),
          'gender': competitor.get('gender', None),
      })

  # Extract sport_event_status
  status = summary['sport_event_status']
  if status.get('status') == 'closed':
    sport_event_status_data.append({
        'event_id': summary['sport_event']['id'],
        'round' : summary['sport_event']['sport_event_context']['round'].get('name'),
        'season': summary['sport_event']['sport_event_context']['season']['name'],
        'stage' : summary['sport_event']['sport_event_context']['stage']['phase'],
        'match_id': summary['sport_event']['sport_event_context']['competition']['id'],
        'match_start_time': summary['sport_event']['start_time'],
        'status': status.get('status'),
        'match_status': status.get('match_status'),
        'home_score': int(status.get('home_score', 0)),
        'away_score': int(status.get('away_score', 0)),
        'winner_id': status.get('winner_id')
  })

  # Extract teams statistics
  if 'statistics' in summary and summary['statistics'] is not None:
    stats_list = summary['statistics'].get('totals')
    # Check if stats_list is not None before accessing 'competitors'
    if stats_list is not None:
      for stats in stats_list.get('competitors', []):
        if stats.get('name') == 'Liverpool FC':
          statistics_data.append({
            "ball_possession": int(stats['statistics'].get("ball_possession", 0)),
            "cards_given": int(stats['statistics'].get("cards_given", 0)),
            "chances_created": int(stats['statistics'].get("chances_created", 0)),
            "clearances": int(stats['statistics'].get("clearances", 0)),
            "corner_kicks": int(stats['statistics'].get("corner_kicks", 0)),
            "crosses_successful": int(stats['statistics'].get("crosses_successful", 0)),
            "crosses_total": int(stats['statistics'].get("crosses_total", 0)),
            "crosses_unsuccessful": int(stats['statistics'].get("crosses_unsuccessful", 0)),
            "defensive_blocks": int(stats['statistics'].get("defensive_blocks", 0)),
            "diving_saves": int(stats['statistics'].get("diving_saves", 0)),
            "dribbles": int(stats['statistics'].get("dribbles_completed", 0)),
            "fouls" : int(stats['statistics'].get("fouls", 0)),
            "free_kicks": int(stats['statistics'].get("free_kicks", 0)),
            "goal_kicks": int(stats['statistics'].get("goal_kicks", 0)),
            "injuries": int(stats['statistics'].get("injuries", 0)),
            "interceptions": int(stats['statistics'].get("interceptions", 0)),
            "long_passes_successful": int(stats['statistics'].get("long_passes_successful", 0)),
            "long_passes_total": int(stats['statistics'].get("long_passes_total", 0)),
            "long_passes_unsuccessful": int(stats['statistics'].get("long_passes_unsuccessful", 0)),
            "loss_of_possession": int(stats['statistics'].get("loss_of_possession", 0)),
            "red_cards": int(stats['statistics'].get("red_cards", 0)),
            "offsides": int(stats['statistics'].get("offsides", 0)),
            "passes_successful": int(stats['statistics'].get("passes_successful", 0)),
            "passes_total": int(stats['statistics'].get("passes_total", 0)),
            "passes_unsuccessful": int(stats['statistics'].get("passes_unsuccessful", 0)),
            "shots_blocked": int(stats['statistics'].get("shots_blocked", 0)),
            "shots_off_target": int(stats['statistics'].get("shots_off_target", 0)),
            "shots_on_target": int(stats['statistics'].get("shots_on_target", 0)),
            "shots_saved": int(stats['statistics'].get("shots_saved", 0)),
            "shots_total": int(stats['statistics'].get("shots_total", 0)),
            "substitutions": int(stats['statistics'].get("substitutions", 0)),
            "tackles_successful": int(stats['statistics'].get("tackles_successful", 0)),
            "tackles_total": int(stats['statistics'].get("tackles_total", 0)),
            "tackles_unsuccessful": int(stats['statistics'].get("tackles_unsuccessful", 0)),
            "throw_ins": int(stats['statistics'].get("throw_ins", 0)),
            "was_fouled": int(stats['statistics'].get("was_fouled", 0)),
            "yellow_cards": int(stats['statistics'].get("yellow_cards", 0)),
            "yellow_red_cards": int(stats['statistics'].get("yellow_red_cards", 0))
        })
        else:
          statistics_oppenents_data.append({
            'opponent_id': stats.get('id'),
            'opponent_name': stats.get('name'),
            'qualifier': stats.get('qualifier')
        })

# Convert lists to DataFrames
competitors_df = pd.DataFrame(competitors_data)
sport_event_status_df = pd.DataFrame(sport_event_status_data)
statistics_df = pd.DataFrame(statistics_data)
statistics_oppenents_data_df = pd.DataFrame(statistics_oppenents_data)

# Save DataFrames to CSV files
#event_df.to_csv('event_data.csv', index=False)
#competitors_df.to_csv('competitors.csv', index=False)
#sport_event_status_df.to_csv('sport_event_status.csv', index=False)

# Print DataFrames
print("Competitors Data:")
print(competitors_df.head())
print("Sport Event Status Data:")
print(sport_event_status_df.head())
print("Statistics Data:")
print(statistics_df.head())
print("Statistics Opponents Data:")
print(statistics_oppenents_data_df.head())
```
